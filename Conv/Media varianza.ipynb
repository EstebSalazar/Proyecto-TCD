```python
import librosa
import librosa.display
import numpy as np
import os
import statistics 
```

```python
l = os.listdir('Train np')
```

```python
lis=[]
for arch in l:
    if arch.startswith('x_'):
        lis.append(arch)
```

```python
X = np.load('Train np/'+lis[0])
print(X.shape)
for i in range(1,len(lis)):
    x = np.load('Train np/'+lis[i])
    print(x.shape)
    X = np.concatenate((X,x), axis=1)
```
