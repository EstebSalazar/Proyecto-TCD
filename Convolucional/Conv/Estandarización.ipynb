```python
import librosa
import librosa.display
import numpy as np
import os
import statistics 
```

```python
l = os.listdir('Train np')
```

```python
lis=[]
for arch in l:
    if arch.startswith('x_'):
        lis.append(arch)
```

```python
X = np.load('Train np/'+lis[0])
print(X.shape)
for i in range(1,len(lis)):
    x = np.load('Train np/'+lis[i])
    X = np.concatenate((X,x), axis=0)
```

```python
X_t = (X-X.min())/(X.max()-X.min())
```

```python
np.mean(X_t,axis=0)
```

```python
np.save('Res/x_mean.npy', np.mean(X_t,axis=0))
np.save('Res/x_min.npy', np.min(X))
np.save('Res/x_max.npy', np.max(X))
```
